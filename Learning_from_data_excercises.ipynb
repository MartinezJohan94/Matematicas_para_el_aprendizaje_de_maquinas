# Exercise 1.3

The weight update rule in (1.3) has the nice interpretation that it moves in the direction of classifying x(t) correctly. 

a) Show that $$y(t)w^Tx(t)<0$$

[Hint: $x(t)$ is misclassified by $w(t)$.]


b) Show that $$y(t)w^T(t+1)x(t)>y(t)w^Tx(t)$$

c) As far as classifying is concerned, argue that the move from to 
 is a move 'in the right direction'.




#Solución 
a) A partir de la ecuación (1.3) se tiene que $w(t+1)=w(t)+y(t)x(t)$. Suponga que $x(t)$ es clasificado erróneamente, entonces, $y(t)w(t)x(t)\leq0$. Así, la resultado actual de $y(t)$ es erróneo.

Ahora, sea el vector de pesos actualizado $w(t+1)$ y su producto escalar con $x(t)$

\begin{align*}
w(t+1)x(t) &= [w(t) + y(t)x(t)]x(t)\\
&= w(t)x(t) + y(t)x(t)x(t)
\end{align*}

Dado que $x(t)x(t)>0$, entonces $w(t+1)x(t)=w(t)x(t)+y(t)||x(t)||^2$

Nótese que $y(t)||x(t)||^2$ representa un factor de escala positivo de $x(t)$ de manera que afecta directamente la dirección de la actualización de pesos ya que $y(t)||x(t)||^2>0$. De esta manera el vector de pesos se moverá en la dirección de $x(t)$, intentando corregir la clasificación incorrecta.

b) A partir de la ecuación (1.3) se sabe que $w(t+1)=w(t)+y(t)x(t)$, entonces,

\begin{align*}
y(t)w^T(t+1)x(t)&=y(t)\bigl[w(t)+y(t)x(t)\bigr]^Tx(t)\\
&=y(t)\bigl[w^T(t)+y(t)x^T(t)\bigr]x(t)\\
&=y(t)w^T(t)x(t)+y(t)y(t)x^T(t)x(t)\\
&>y(t)w^T(t)x(t)
\end{align*}

c) A partir del literal b) se tiene que $y(t)w^T(t)x(t)$ es creciente. Luego, si $y(t)>0$ y $w^T(t)x(t)< 0$ entonces será posible desplazarse desde $w^T(t)x(t)$ hacia valores positivos. 
De manera similar, si $y(t)< 0$ y $w^T(t)x(t)>0$ entonces se tendrá que $y(t)w^T(t)x(t)$ estará disminuyendo de manera que se desplaza hacia valores negativos. 

Así, cuando se realiza un desplazamiento de $w(t)$ a $w(t+1)$ este se considerará correcto respecto a la clasificación de $x(t)$.
